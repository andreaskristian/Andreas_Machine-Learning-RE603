## Object Oriented Programming (OOP) di Python

Object Oriented Programming atau yang biasa disingkat dengan OOP adalah suatu paradigma pemograman yang berfokus pada objek. Dalam pemograman OOP, kita memandang suatu program berisikan objek-objek yang saling mendukung dan berinteraksi satu sama lain. Tidak semua bahasa pemograman dapat mendukung paradigma pemograman ini dan Python adalah salah satu bahasa pemograman yang mendukung OOP.

## Enkapsulasi

```python
class User:
    def __init__(self, username, email, gender):
        self._username = username  # Enkapsulasi dengan menggunakan underscore (_)
        self._email = email
        self._gender = gender

    def get_username(self):
        return self._username

    def set_username(self, username):
        # Validasi atau logika tambahan dapat ditambahkan di sini
        self._username = username
```

```python
# Contoh penggunaan
user = User("Jonathan Adis Pratama", "Jonathan@example.com", "Male")
print(user.get_username())  # Mengakses melalui metode getter
user.set_username("Jo saja")  # Mengubah melalui metode setter
print(user.get_username())
```

1. Enkapsulasi digunakan untuk menyembunyikan implementasi internal dan melindungi atribut agar tidak dapat diakses atau diubah secara langsung.
2. Atribut username, email, dan gender diberi awalan _ untuk menandakan bahwa seharusnya diakses secara terbatas.
3. Metode getter get_username digunakan untuk mengakses nilai atribut, dan setter set_username digunakan untuk mengubahnya dengan validasi atau logika tambahan.

## Inheritance

Dalam OOP, dikenal suatu istilah yang bernama inheritance atau pewarisan. Istilah ini mengisyaratkan bahwa kita dapat membuat suatu kelas baru yang berasal dari turunan kelas yang lain.

Kelas yang merupakan turunan (atau children/anak) dari kelas lainnya (biasa disebut dengan parent) akan mewarisi attribut dan method yang berasal dari parentnya.

Child class juga dapat melakukan overriding terhadap method yang terdapat dari parent class.

```python
class PremiumUser(User):
    def __init__(self, username, email, gender, premium_level):
        super().__init__(username, email, gender)
        self.premium_level = premium_level

    def display_premium_info(self):
        print("Premium Level:", self.premium_level)
```

```python
premium_user = PremiumUser("Jonathan Adis Pratama", "Jonathan@example.com", "Male", "Gold")
premium_user.display_premium_info()
```

1. Inheritance memungkinkan kelas anak (PremiumUser) untuk mewarisi atribut dan metode dari kelas induk (User).
2. Melalui super().__init__, konstruktor kelas anak memanggil konstruktor kelas induk untuk menginisialisasi atribut yang sama.
3. Kelas anak dapat menambahkan metode tambahan seperti display_premium_info

## Polymorphism

```python
class User:
    def __init__(self, username, email, gender):
        self.username = username
        self.email = email
        self.gender = gender

    def display_user_info(self):
        print("Username:", self.username)
        print("Email:", self.email)
        print("Gender:", self.gender)

class PremiumUser(User):
    def __init__(self, username, email, gender, premium_level):
        super().__init__(username, email, gender)
        self.premium_level = premium_level

    def display_user_info(self):  # Override metode display_user_info dari kelas User
        super().display_user_info()  # Memanggil metode display_user_info dari kelas induk
        print("Premium Level:", self.premium_level)
```

```python
def display_user_information(user):
    if isinstance(user, User):
        user.display_user_info()
    else:
        print("Objek tidak dikenali.")
```

```python
# Contoh penggunaan
user1 = User("Jonathan Adis Pratama", "Jonathan@example.com", "Male")
user2 = PremiumUser("Jonathan Adis Pratama", "Jonathan@example.com", "Male", "Gold",)

display_user_information(user1)  # Memanggil metode display_user_info dari kelas User
display_user_information(user2)  # Memanggil metode display_user_info dari kelas PremiumUser
```

Dalam contoh code, display_user_information berperan seperti remote kontrol tersebut. Meskipun kita mungkin memiliki objek dari kelas User dan PremiumUser, fungsi ini mampu "mengendalikan" keduanya dengan cara yang sama, yaitu memanggil metode display_user_info. Hal ini mencerminkan prinsip polimorfisme di mana suatu fungsi dapat bekerja dengan objek-objek yang berbeda dengan menggunakan antarmuka yang seragam.

Dengan polimorfisme, kita dapat menulis kode yang lebih fleksibel dan mudah diubah, karena tidak perlu tahu tipe objek secara spesifik, tetapi cukup tahu bahwa objek tersebut memiliki antarmuka yang diperlukan.

## Abstraction

```python
import abc
print(dir(abc))
```

* *ABC* digunakan untuk membuat kelas abstrak, yaitu kelas yang tidak bisa langsung dibuat objeknya.
* *abstractmethod* digunakan untuk mendeklarasikan method abstrak, yaitu method yang harus diimplementasikan oleh subclass.

```python
from abc import ABC, abstractmethod

# Kelas abstrak (abstraksi) untuk jenis mobil
class JenisMobil(ABC): #kelas yang diwarisi oleh ABC
    def __init__(self, nama, warna):
        self.nama = nama
        self.warna = warna

    @abstractmethod
    def informasi(self):
        pass

# Kelas turunan yang mengimplementasikan metode abstrak dari JenisMobil
class Sedan(JenisMobil):
    def informasi(self):
        return f"Sedan {self.nama} berwarna {self.warna}"

# Kelas turunan lainnya
class SUV(JenisMobil):
    def informasi(self):
        return f"SUV {self.nama} berwarna {self.warna}"

# Fungsi untuk menampilkan informasi mobil
def tampilkan_informasi(mobil):
    print(mobil.informasi())
```

```python
# Contoh penggunaan
sedan_mobil = Sedan("Civic", "Merah")
suv_mobil = SUV("CR-V", "Hitam")

tampilkan_informasi(sedan_mobil)
tampilkan_informasi(suv_mobil)
```

### A. Abstraksi:

    - Kelas abstrak JenisMobil menyediakan kerangka kerja untuk jenis mobil apa pun. Atribut seperti nama dan warna adalah aspek-aspek abstrak yang relevan untuk semua jenis mobil.
    - Metode abstrak informasi memungkinkan setiap kelas turunan untuk memberikan implementasi khususnya sesuai dengan jenis mobil yang mereka wakili.

### B. Penggunaan Abstraksi:

    - Penggunaan fungsi tampilkan_informasi menunjukkan bagaimana kita dapat menggunakan antarmuka umum (informasi) untuk menampilkan informasi mobil tanpa perlu tahu detail implementasinya.
    - Pengguna luar hanya perlu tahu bahwa objek dapat memberikan informasi tentang jenis mobil, dan detail implementasi dikurangi.

### Program Sistem Tanaman
Program ini mendemonstrasikan 3 prinsip OOP: **Enkapsulasi**, **Inheritance**, dan **Polymorphism**

```python
from abc import ABC, abstractmethod

# ===== 1. ENKAPSULASI =====
# Kelas dasar Tanaman dengan atribut yang di-enkapsulasi
class Tanaman:
    def __init__(self, nama, jenis, umur_bulan):
        self._nama = nama  # Atribut protected (enkapsulasi)
        self._jenis = jenis
        self._umur_bulan = umur_bulan
        self._sudah_ditanam = False  # Status apakah sudah ditanam
        self._kondisi_sehat = "Baik"

    # Getter methods
    def get_nama(self):
        return self._nama

    def get_jenis(self):
        return self._jenis

    def get_umur_bulan(self):
        return self._umur_bulan

    def is_ditanam(self):
        return self._sudah_ditanam

    def get_kondisi_sehat(self):
        return self._kondisi_sehat

    # Setter methods dengan validasi
    def set_status_tanam(self, status):
        if isinstance(status, bool):
            self._sudah_ditanam = status
        else:
            print("Status penanaman harus boolean (True/False)")

    def set_kondisi_sehat(self, kondisi):
        valid_kondisi = ["Baik", "Sedang", "Buruk"]
        if kondisi in valid_kondisi:
            self._kondisi_sehat = kondisi
        else:
            print(f"Kondisi '{kondisi}' tidak valid. Gunakan: {', '.join(valid_kondisi)}")

    # Method umum untuk menampilkan info
    def tampilkan_info(self):
        status = "Sudah Ditanam" if self._sudah_ditanam else "Belum Ditanam"
        print(f"Nama Tanaman: {self._nama}")
        print(f"Jenis: {self._jenis}")
        print(f"Umur: {self._umur_bulan} bulan")
        print(f"Status Penanaman: {status}")
        print(f"Kondisi Kesehatan: {self._kondisi_sehat}")


# ===== 2. INHERITANCE =====
# Kelas turunan untuk pohon buah
class PohonBuah(Tanaman):
    def __init__(self, nama, jenis, umur_bulan, musim_panen, hasil_per_musim):
        super().__init__(nama, jenis, umur_bulan)  # Mewarisi dari parent
        self._musim_panen = musim_panen
        self._hasil_per_musim = hasil_per_musim  # kg, buah, dll.

    def get_musim_panen(self):
        return self._musim_panen

    def get_hasil_per_musim(self):
        return self._hasil_per_musim

    # Override method tampilkan_info (Polymorphism)
    def tampilkan_info(self):
        super().tampilkan_info()  # Panggil method parent
        print(f"Musim Panen: {self._musim_panen}")
        print(f"Potensi Hasil per Musim: {self._hasil_per_musim}")


# Kelas turunan untuk pohon hias
class PohonHias(Tanaman):
    def __init__(self, nama, jenis, umur_bulan, warna_bunga, tinggi_maksimal_cm):
        super().__init__(nama, jenis, umur_bulan)
        self._warna_bunga = warna_bunga
        self._tinggi_maksimal_cm = tinggi_maksimal_cm

    def get_warna_bunga(self):
        return self._warna_bunga

    def get_tinggi_maksimal_cm(self):
        return self._tinggi_maksimal_cm

    # Override method tampilkan_info (Polymorphism)
    def tampilkan_info(self):
        super().tampilkan_info()
        print(f"Warna Bunga: {self._warna_bunga}")
        print(f"Tinggi Maksimal: {self._tinggi_maksimal_cm} cm")


# ===== 3. POLYMORPHISM =====
# Fungsi yang dapat menerima berbagai jenis tanaman
def tanam_pohon(tanaman):
    """Fungsi polymorphic yang bisa memproses semua jenis tanaman untuk ditanam"""
    print("\n" + "="*50)
    print("PROSES PENANAMAN POHON")
    print("="*50)

    if not tanaman.is_ditanam():
        tanaman.set_status_tanam(True)
        print("✓ Penanaman berhasil!")
        print("\nDetail Tanaman yang Ditanam:")
        tanaman.tampilkan_info()  # Polymorphic call
    else:
        print("✗ Tanaman sudah ditanam sebelumnya.")
        tanaman.tampilkan_info()


def rawat_pohon(tanaman, kondisi):
    """Fungsi untuk merawat pohon dan mengubah kondisinya"""
    print("\n" + "="*50)
    print("PERAWATAN POHON")
    print("="*50)

    if tanaman.is_ditanam():
        tanaman.set_kondisi_sehat(kondisi)
        print(f"✓ Perawatan pohon '{tanaman.get_nama()}' berhasil. Kondisi: {tanaman.get_kondisi_sehat()}")
        tanaman.tampilkan_info()
    else:
        print(f"✗ Pohon '{tanaman.get_nama()}' belum ditanam, tidak bisa dirawat.")
```

```python
# Membuat objek Pohon Buah
pohon_mangga = PohonBuah(
    "Mangga Harum Manis",
    "Buah Tropis",
    24, # umur 24 bulan
    "Musim Hujan",
    "50 kg per musim"
)

# Membuat objek Pohon Hias
pohon_mawar = PohonHias(
    "Mawar Merah",
    "Hias",
    6, # umur 6 bulan
    "Merah",
    150 # tinggi maksimal 150 cm
)

# Demonstrasi Enkapsulasi - Akses menggunakan Getter
print("="*50)
print("CONTOH ENKAPSULASI - Akses menggunakan Getter")
print("="*50)
print(f"Nama Pohon Buah: {pohon_mangga.get_nama()}")
print(f"Jenis: {pohon_mangga.get_jenis()}")
print(f"Musim Panen: {pohon_mangga.get_musim_panen()}")

print(f"\nNama Pohon Hias: {pohon_mawar.get_nama()}")
print(f"Warna Bunga: {pohon_mawar.get_warna_bunga()}")

# Demonstrasi Inheritance & Polymorphism melalui fungsi tanam_pohon
print("\n" + "="*50)
print("CONTOH INHERITANCE & POLYMORPHISM (Fungsi tanam_pohon)")
print("="*50)

tanam_pohon(pohon_mangga) # Proses penanaman pohon mangga
tanam_pohon(pohon_mawar)  # Proses penanaman pohon mawar

# Coba tanam pohon yang sudah ditanam
tanam_pohon(pohon_mangga)

# Demonstrasi Polymorphism melalui fungsi rawat_pohon
print("\n" + "="*50)
print("CONTOH POLYMORPHISM (Fungsi rawat_pohon)")
print("="*50)

rawat_pohon(pohon_mangga, "Sedang") # Ubah kondisi pohon mangga
rawat_pohon(pohon_mawar, "Buruk")   # Ubah kondisi pohon mawar
rawat_pohon(pohon_mangga, "Baik")   # Kembalikan kondisi pohon mangga ke Baik
```

### Penjelasan Prinsip OOP yang Diterapkan:

**1. Enkapsulasi:**
- Atribut class `Tanaman` menggunakan underscore `_` (protected attributes), seperti `_nama`, `_jenis`, `_umur_bulan`, `_sudah_ditanam`, dan `_kondisi_sehat`. Ini bertujuan untuk menyembunyikan detail implementasi internal dari akses langsung.
- Akses dan modifikasi atribut-atribut ini dilakukan melalui metode *getter* (`get_nama()`, `is_ditanam()`, dll.) dan *setter* (`set_status_tanam()`, `set_kondisi_sehat()`).
- Metode setter `set_status_tanam()` dan `set_kondisi_sehat()` memiliki validasi untuk memastikan input yang diberikan benar dan sesuai (misalnya, `status` harus boolean, `kondisi` harus salah satu dari `"Baik", "Sedang", "Buruk"`). Ini membantu menjaga integritas data objek.

**2. Inheritance (Pewarisan):**
- Class `PohonBuah` dan `PohonHias` adalah class anak (child classes) yang mewarisi sifat dan perilaku dari class `Tanaman` (parent class). Ini terlihat dari deklarasi `class PohonBuah(Tanaman):` dan `class PohonHias(Tanaman):`.
- Menggunakan `super().__init__(nama, jenis, umur_bulan)` di konstruktor class anak memungkinkan mereka memanggil konstruktor class parent, sehingga atribut dasar seperti nama, jenis, dan umur tanaman diinisialisasi oleh class `Tanaman`.
- Masing-masing class anak kemudian menambahkan atribut dan metode spesifik yang relevan dengan jenisnya, misalnya:
    - `PohonBuah` menambahkan atribut `_musim_panen` dan `_hasil_per_musim`, beserta metode getter-nya.
    - `PohonHias` menambahkan atribut `_warna_bunga` dan `_tinggi_maksimal_cm`, beserta metode getter-nya.

**3. Polymorphism:**
- Metode `tampilkan_info()` di-override (ditimpa) pada setiap class anak (`PohonBuah` dan `PohonHias`). Artinya, meskipun nama metodenya sama, implementasinya berbeda di setiap class anak untuk menampilkan informasi yang spesifik dan relevan dengan jenis pohon tersebut.
- Fungsi `tanam_pohon(tanaman)` dan `rawat_pohon(tanaman, kondisi)` adalah contoh polymorphism. Kedua fungsi ini dirancang untuk dapat menerima objek dari class `Tanaman`, `PohonBuah`, atau `PohonHias` sebagai argumen `tanaman`.
- Saat `tanaman.tampilkan_info()` dipanggil di dalam fungsi-fungsi ini, Python secara otomatis akan memanggil implementasi `tampilkan_info()` yang benar sesuai dengan tipe objek aktual yang diteruskan (baik itu `PohonBuah` atau `PohonHias`). Hal ini menghasilkan kode yang lebih fleksibel dan mudah diperluas, karena satu antarmuka (`tampilkan_info()`) dapat menghasilkan perilaku yang berbeda tergantung pada tipe objek yang memanggilnya.
